```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_fwf('auto-mpg.data', header=None)
df
```

```python
df.columns =[
    'mpg', 
    'cylinders', 
    'displacement', 
    'horsepower', 
    'weight', 
    'acceleration', 
    'model year', 
    'origin', 
    'car name']

df
```

```python
data = np.where(df['horsepower'] == '?')
data
```

```python
df['horsepower'].loc[[32, 126, 330, 336, 354, 374]] = np.nan
```

```python
df.iloc[[32, 126, 330, 336, 354, 374]]
```

```python
df.dtypes
```

```python
df['horsepower'] = df['horsepower'].astype(float)
```

```python
usa = np.where(df['origin'] == 1)
asia = np.where(df.origin == 2)
europe = np.where(df.origin == 3)

df.origin.loc[usa] = 'USA'
df.origin.loc[asia] = 'Asia'
df.origin.loc[europe] = 'Europe'

df
```

```python
sns.barplot(data=df, x='cylinders', y='mpg')
```

```python
sns.scatterplot(data=df, x='weight', y='horsepower', hue='mpg')
```
